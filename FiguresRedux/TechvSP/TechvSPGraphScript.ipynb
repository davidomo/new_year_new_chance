```python
# Dependencies
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "../../Figures/SP500/SPDaymasterdata.csv"
SPDay = pd.read_csv(file_name)
file_name2 = "../../Figures/SP500/SPMonthmasterdata.csv"
SPMonth = pd.read_csv(file_name2)
file_name3 = "../../Figures/Tech/TechDaymasterdata.csv"
TechDay = pd.read_csv(file_name3)
file_name4 = "../../Figures/Tech/TechMonthmasterdata.csv"
TechMonth = pd.read_csv(file_name4)
```

```python
TechMon
```

```python
Days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
Months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
x_axis = np.arange(len(Days))
Tech = TechDay["Daily Change Average"]
SP = SPDay["Daily Change Average"]

width = .35

plt.figure(figsize=(10,8))

plt.bar(x_axis, SP, width, alpha=0.8, align="edge", label = "S&P 500")
plt.bar(x_axis + width, Tech, width, alpha=0.8, align="edge", label = "Tech")

plt.axhline(y=0, linestyle='-', linewidth = 1, color = "black", alpha = .6)

tick_locations = [value+width for value in x_axis]
plt.xticks(tick_locations, Days)
plt.legend()

plt.ylabel("Average of Open/Close Change (%)")
plt.title("Tech vs. S&P 500 Daily Price Change")

plt.savefig("TechvSPDDCA")
plt.show()
```
