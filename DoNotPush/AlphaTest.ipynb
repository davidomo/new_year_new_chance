```python
mydictionary = getJ
df = pd.DataFrame.from_dict(mydictionary["Monthly Time Series"], orient='index')
df.head()
```

```python
df = df.reset_index()
df = df.sort_index(ascending=False)
df = df.reset_index()
df.head()
```

```python
month = []
year = []

for index, row in df.iterrows():
    a = row["index"]
    datee = datetime.strptime(a, "%Y-%m-%d")
    monthdate = datee.strftime("%B")
    yeardate = datee.strftime("%Y")
    month.append(monthdate)
    year.append(yeardate)

df["month"] = month
df["year"] = year

df.head()
```

```python
# New - Old / Old

change = []

counter = 0
while counter < (len(df["index"]) -1) :
    new = df["4. close"][counter]
    old = df["4. close"][counter + 1]
    calc = (float(new) - float(old)) / float(old)
    change.append(calc)
    counter += 1

change.append(0)
df["month change"] = change
df.head()
```
