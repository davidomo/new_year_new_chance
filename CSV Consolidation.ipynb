```python
# Import Dependencies
import pandas as pd

AAPL_df = pd.read_csv("JNJfinal.CSV")
AMZN_df = pd.read_csv("MRKfinal.CSV")
FB_df = pd.read_csv("PFEfinal.CSV")
GOOGL_df = pd.read_csv("NVSfinal.CSV")
MSFT_df = pd.read_csv("ROGfinal.CSV")

AAPL_df.head()
```

```python
Master_df = pd.concat([AAPL_df,AMZN_df,FB_df,GOOGL_df,MSFT_df], axis=0)

Change = []
counter = 0
while counter < len(Master_df["4. close"]):
    o = Master_df["1. open"][counter]
    c = Master_df["4. close"][counter]
    h = (o - c) / c
    Change.append(h)
    counter += 1
    
Change
    
```

```python
Master_df.to_csv("Pharm_masterdata.csv", index=False, header=True)
```
