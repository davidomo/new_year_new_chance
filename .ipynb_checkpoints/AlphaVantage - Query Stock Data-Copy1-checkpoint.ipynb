```python
import requests
import json
import pprint
import pandas as pd
from datetime import datetime

url = "https://www.alphavantage.co/query"

function = "TIME_SERIES_MONTHLY"
symbol = "MSFT"
api_key = "SMBQT4NW44MXXREK"   

apiurl = "%s?function=%s&symbol=%s&api_key=%s" % (url,function,symbol,api_key)
print(apiurl)

```

```python
data = { "function": function,
         "symbol": symbol,
         "apikey": api_key } 
page = requests.get(url, params = data)

pprint.pprint(page.json())
```

```python
mydictionary = page.json()
df = pd.DataFrame.from_dict(mydictionary["Monthly Time Series"], orient='index')
```

```python
print(df.iloc[1:10])
```

```python
df = df.reset_index()
df = df.sort_index(ascending=False)
df.head()
```

```python
month = []
for index, row in df.iterrows():
    a = row["index"]
    datee = datetime.strptime(a, "%Y-%m-%d")
    monthdate = datee.strftime("%B")
    month.append(monthdate)
print(month)
```

```python
df["Month"] = month
df.head()
```
