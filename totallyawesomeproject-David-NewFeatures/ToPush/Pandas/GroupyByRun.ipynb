```python
# Group by day of week
gAAPL = AAPL_df.groupby(["Day"])

gDailyChange = gAAPL["daily change"].mean()
gVolume = gAAPL["5. volume"].sum()

printable_df = pd.DataFrame({"Daily Change Avg": gDailyChange,
                             "Daily Volume": gVolume})
printable_df = printable_df.reset_index()
```

```python
day_to_num = []

x = 0

while x < len(master_df["Day"]):
    
    if master_df["Day"][x] == "Monday":
        day_to_num.append(1)
    elif master_df["Day"][x] == "Tuesday":
        day_to_num.append(2)
    elif master_df["Day"][x] == "Wednesday":
        day_to_num.append(3)
    elif master_df["Day"][x] == "Thursday":
        day_to_num.append(4)
    else:
        day_to_num.append(5)
        
    x += 1

master_df["Date to Num"] = day_to_num
master_df.head(10)
```
