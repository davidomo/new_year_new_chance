```python
# Dependencies
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
file_name = "S&P.csv"
master_df = pd.read_csv(file_name)
master_df.head()
```

```python
master_df["Date"]=master_df["Date"].astype(str)
# master_df.dropna(how='all')
master_df.head()
```

```python
month = []
day=[]
year = []

x = 0

while x < len(master_df["Date"]):
    date = datetime.strptime(master_df["Date"][x], "%m/%d/%Y")
    monthdate = date.strftime("%B")
    daydate = date.strftime("%A") 
    yeardate = date.strftime("%Y") 
    month.append(monthdate)
    year.append(yeardate)
    day.append(daydate) 
    x += 1
    

master_df["Month"] = month
master_df["Day"] = day
master_df["Year"]= year
master_df.head()


```

```python
master_df.to_csv("SPreworked.csv")
```
