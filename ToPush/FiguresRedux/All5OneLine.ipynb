```python
# Dependencies
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "../Figures/SP500/SPDaymasterdata.csv"
SPDay = pd.read_csv(file_name)
file_name2 = "../Figures/SP500/SPMonthmasterdata.csv"
SPMonth = pd.read_csv(file_name2)
file_name3 = "../Figures/Tech/Redux/TechDaymasterdata2.csv"
TechDay = pd.read_csv(file_name3)
file_name4 = "../Figures/Tech/Redux/TechMonthmasterdata2.csv"
TechMonth = pd.read_csv(file_name4)
file_name5 = "../Figures/Vidya/Redux/GameDaymasterdata2.csv"
GameDay = pd.read_csv(file_name5)
file_name6 = "../Figures/Vidya/Redux/GameMonthmasterdata2.csv"
GameMonth = pd.read_csv(file_name6)
file_name7 = "../Figures/Pharma/PharmaDaymasterdata2.csv"
PharDay = pd.read_csv(file_name7)
file_name8 = "../Figures/Pharma/PharmaMonthmasterdata2.csv"
PharMonth = pd.read_csv(file_name8)
file_name9 = "../Figures/Food/FoodDaymasterdata2.csv"
FoodDay = pd.read_csv(file_name9)
file_name10 = "../Figures/Food/FoodMonthmasterdata2.csv"
FoodMonth = pd.read_csv(file_name10)
file_name11 = "../Figures/Finance/Redux/FinanceDaymasterdata2.csv"
BankDay = pd.read_csv(file_name11)
file_name12 = "../Figures/Finance/Redux/FinanceMonthmasterdata2.csv"
BankMonth = pd.read_csv(file_name12)
```

```python
Days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
Months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
x_axis = np.arange(len(Days))
Tech = TechDay["Daily Change Average"]
SP = SPDay["Daily Change Average"]
Game = GameDay["Daily Change Average"]
Phar = PharDay["Daily Change Average"]
Food = FoodDay["Daily Change Average"]
Bank = BankDay["Daily Change Average"]

plt.figure(figsize=(12,8))

plt.plot(x_axis, SP, alpha=0.8, label = "S&P 500", linewidth = 5, color = "black")
plt.plot(x_axis, Tech, alpha=0.8, label = "Tech", linewidth = 5)
plt.plot(x_axis, Game, alpha=0.8, label = "Game", linewidth = 5)
plt.plot(x_axis, Phar, alpha=0.8, label = "Pharma", linewidth = 5)
plt.plot(x_axis, Food, alpha=0.8, label = "Food", linewidth = 5)
plt.plot(x_axis, Bank, alpha=0.8, label = "Bank", linewidth = 5)

plt.axhline(y=0, linestyle='-', linewidth = 1, color = "black", alpha = .6)

z = 1

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, Days)
plt.legend()

plt.ylabel("Average of Open/Close Change (%)")
plt.title("All Groups vs. S&P 500 Daily Price Change")

plt.savefig("AllvSPDDCA")
plt.show()
```
